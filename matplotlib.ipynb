# BAR PLOT

```python
plt.bar?
```

```python
plt.bar([0,1,2], [10,20,15])
```

```python
x_cor = np.array([0,1,2])
plt.bar(x_cor, [10,20,15], width = 0.5, label = 'Aqua data', color = 'aqua')
plt.bar(x_cor + 0.5, [20,10,13], width = 0.5, color = 'red', label = 'Red data', tick_label = ['x', 'y', 'z'])
plt.ylim(0, 30)
plt.xlim(-1, 3.5)
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()
plt.title('Adorable Bar')
```

```python
x_gaussian = np.random.randn(1000)
sigma = 5
mean = 60
x_shifted_gaussian_1 = x_gaussian*sigma + mean
x_shifted_gaussian_1 = np.round(x_shifted_gaussian_1)
x_shifted_gaussian_2 = x_gaussian*6 + 40
x_shifted_gaussian_2 = np.round(x_shifted_gaussian_2)
plt.hist(x_shifted_gaussian_1)
plt.hist(x_shifted_gaussian_2, alpha = 0.5)
plt.show()
```
