```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset.keys()
```

```python
print(dataset['target_names'])
print(dataset['feature_names'])
print(dataset['feature_names'].shape)
```

```python
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
```

```python
scale = StandardScaler()
scale.fit(df)
```

```python
scaled_data = scale.fit_transform(df)
```

```python
scaled_data.shape
```

```python
pca = PCA(n_components=3)
pca.fit(scaled_data)
X_reduced = pca.fit_transform(scaled_data)
```

```python
X_reduced.shape
```

```python
pca.components_
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c = dataset['target'])
```

```python
ax = plt.subplot(projection='3d')
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c = dataset['target'])
```
